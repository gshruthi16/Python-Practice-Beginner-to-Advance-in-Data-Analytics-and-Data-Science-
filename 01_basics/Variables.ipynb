### Q14. Create variables for the 3 embarkation towns and their passenger counts:

Southampton → 644, 
Cherbourg → 168, 
Queenstown → 77, 
Calculate total and what percentage came from Southampton. Print formatted to 2 decimal places.

```python
Southampton = 644
Cherbourg = 168
Queenstown = 77

total = Southampton + Cherbourg + Queenstown

percentage = (Southampton / total)*100

print("Total = ", total)
print(f"Percentage came from Southampton = {percentage:.2f}" )
```
